```python
os.mkdir('avaliacoes_derm')
# os.mkdir('avaliacoes_clinica/sebo')
# os.mkdir('avaliacoes_clinica/nevo')
os.mkdir('avaliacoes_derm/melanoma')
# os.mkdir('avaliacoes_derm/cec')
```

```python
def change_img_id(row):
    image = row['img-id'].split('/')[-1]
    path = '/datasets/PAD-UFES-26/anonymous-images/' + image
    return path
```

```python
df = pd.read_csv('data/pad-ufes-26-diag_dermatoscope_folders_raw.csv')
df['img-id'].iloc[0]
```

```python
df['img-id'] = df.apply(change_img_id, axis=1)
```

```python
def copy_images(row):
    # if row['histoMacroCID'] == 'SEBO':
    #     dest_folder = 'avaliacoes_clinica/sebo/'
    # elif row['histoMacroCID'] == 'NEVO':
    #     dest_folder = 'avaliacoes_clinica/nevo/'
    if row['histoMacroCID'] == 'MEL':
        dest_folder = 'avaliacoes_derm/melanoma/'
    else:
        return
    
    src_path = row['img-id']
    dest_path = dest_folder
    os.system(f'cp {src_path} {dest_path}')
                            
```

```python
df.apply(copy_images, axis=1)
```
